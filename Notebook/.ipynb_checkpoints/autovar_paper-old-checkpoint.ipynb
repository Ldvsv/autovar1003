```python
# Packages
import pygeostat as gs
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
import warnings 
warnings.simplefilter('ignore')
import seaborn as sns
import matplotlib.gridspec as gridspec
from sklearn.metrics import mean_squared_error
import sys
import subprocess
import time
print (' ')
print ('Packages versions:')
print (' ')
print ('Pygeostat version :', gs.__version__)
print ('Numpy version     :', np.__version__)
print ('Pandas version    :', pd.__version__)
mpl.style.use('default')

# Directories
datdir = './data/'
exedir = './exes/'
outdir = 'outputs/'
figdir = 'figures/'

gs.mkdir(outdir)
gs.mkdir(figdir)
```

```python
# defining plotting function for Geolocated variogram 2D
def plot_autovar2d(model,experimental,debug,name,xlimmaj,xlimmin,xlimter,title):
    
    model = model
    experimental = experimental
    debug = debug
    fig, axes = plt.subplots(1,2, figsize = (8,4))
    axes[-1].axis('off')
    # Variogram plots
    gs.variogram_plot(model, index = 1, experimental=False, ax = axes[0],color = 'blue',lw =2 , 
                               label = 'Major dir', plot_style = 'mpldefault' )
    gs.variogram_plot(model, index = 2, experimental=False, ax = axes[0],color = 'green' ,lw =2, 
                               label = 'Minor dir', plot_style = 'mpldefault' )

    gs.variogram_plot(experimental, index = 1, experimental=True, ax = axes[0], color = 'blue',ylim = (0,1.8),
                      xlim=(0,xlimmaj),label = 'Major dir', plot_style = 'mpldefault' )
    gs.variogram_plot(experimental, index = 2, experimental=True, ax = axes[0], color = 'green',ylim = (0,1.8),
                      xlim=(0,xlimmin),label = 'Minor dir', plot_style = 'mpldefault' )
    
    axes[0].set_ylabel(r'$\gamma$   ',fontsize =18)
    axes[0].legend(loc='best',fontsize=11)
    axes[0].tick_params(axis='both', labelsize=12)
    # Directions plot

    x, y, z = [0, debug['x y z'][0]], [0, debug['x y z'][3]], [0, debug['x y z'][6]]
    x2, y2, z2 = [0, debug['x y z'][1]], [0, debug['x y z'][4]], [0, debug['x y z'][7]]
    x3, y3, z3 = [0, debug['x y z'][2]], [0, debug['x y z'][5]], [0, debug['x y z'][8]] 

    axes[-1] = fig.add_subplot(133, projection='3d')

    axes[-1].scatter(x, y, z, c='blue', s=100)
    axes[-1].scatter(x2, y2, z2, c='green', s=100)
    axes[-1].scatter(x3, y3, z3, c='red', s=100)

    axes[-1].set_xlabel('X', fontsize=20, rotation=0, labelpad = 15)
    axes[-1].set_ylabel('Y', fontsize=20,)
    axes[-1].set_zlabel('Z', fontsize=20)
    axes[-1].set_zticks([1])
    labels = [-1,-0.5,0,0.5,1]
    labelss = [1]
    axes[-1].set_xticklabels(labels, fontsize = 15)
    axes[-1].set_yticklabels(labels, fontsize = 15)
    axes[-1].set_zticklabels(labelss, fontsize = 15)
    axes[-1].xaxis.set_ticks_position('top')
    axes[-1].zaxis.set_ticks_position('top')


    axes[-1].plot(x, y, z, color='blue')
    axes[-1].plot(x2, y2, z2, color='green')
    axes[-1].plot(x3, y3, z3, color='red')
    axes[-1].view_init(elev=90, azim=-90)
    axes[-1].set_xlim([-1.2, 1.2])
    axes[-1].set_ylim([-1.2, 1.2])
    axes[-1].set_zlim([-1.2, 1.2])
    
#     plt.suptitle(title, fontsize =16,y=1.05)
    plt.tight_layout()
    plt.subplots_adjust(wspace = 0.2)
    gs.export_image( 'figures/{}.png'.format(name), dpi =300)
```

```python
# # defining plotting function For 2-D variograms / Without using the geolocated option
def plot_autovar(model,experimental,debug,name,xlimmaj,xlimmin,xlimter,title):
    
    model = model
    experimental = experimental
    debug = debug
    fig, axes = plt.subplots(1,3, figsize = (12,4))
    axes[-1].axis('off')
    # Variogram plots
    gs.variogram_plot(model, index = 1, experimental=False, ax = axes[0],color = 'blue',lw =2 , 
                               label = 'Major dir')
    gs.variogram_plot(model, index = 2, experimental=False, ax = axes[0],color = 'green' ,lw =2, 
                               label = 'Minor dir')
    gs.variogram_plot(model, index = 3, experimental=False, ax = axes[1],color = 'red',lw =2 , 
                                 label = 'Tert. dir')

    gs.variogram_plot(experimental, index = 1, experimental=True, ax = axes[0], color = 'blue',ylim = (0,1.8),
                      xlim=(0,xlimmaj),label = 'Major dir')
    gs.variogram_plot(experimental, index = 2, experimental=True, ax = axes[0], color = 'green',ylim = (0,1.8),
                      xlim=(0,xlimmin),label = 'Minor dir' )
    gs.variogram_plot(experimental, index = 3, experimental=True, ax = axes[1], color = 'red',ylim = (0,1.8),
                      xlim=(0,xlimter),label = 'Tert. dir')
    
    axes[0].set_ylabel(r'$\gamma$   ',fontsize =18)
    axes[1].set_ylabel(r'$\gamma$   ',fontsize =18)
    axes[0].tick_params(axis='both', labelsize=12)
    axes[1].tick_params(axis='both', labelsize=12)
    axes[0].legend(loc='best', fontsize = 11)
    axes[1].legend(loc='best', fontsize = 11)
    # Directions plot

    x, y, z = [0, debug['x y z'][0]], [0, debug['x y z'][3]], [0, debug['x y z'][6]]
    x2, y2, z2 = [0, debug['x y z'][1]], [0, debug['x y z'][4]], [0, debug['x y z'][7]]
    x3, y3, z3 = [0, debug['x y z'][2]], [0, debug['x y z'][5]], [0, debug['x y z'][8]] 

    axes[-1] = fig.add_subplot(133, projection='3d')

    axes[-1].scatter(x, y, z, c='blue', s=100)
    axes[-1].scatter(x2, y2, z2, c='green', s=100)
    axes[-1].scatter(x3, y3, z3, c='red', s=100)

    axes[-1].set_xlabel('X', fontsize=17, rotation=0, labelpad = 15)
    axes[-1].set_ylabel('Y', fontsize=17,)
    axes[-1].set_zlabel('Z', fontsize=17)
    axes[-1].set_zticks([1])
    labels = [-1,-0.5,0,0.5,1]
    labelss = [1]
    axes[-1].set_xticklabels(labels, fontsize = 15)
    axes[-1].set_yticklabels(labels, fontsize = 15)
    axes[-1].set_zticklabels(labelss, fontsize = 15)
    axes[-1].xaxis.set_ticks_position('top')
    axes[-1].zaxis.set_ticks_position('top')


    axes[-1].plot(x, y, z, color='blue')
    axes[-1].plot(x2, y2, z2, color='green')
    axes[-1].plot(x3, y3, z3, color='red')
    axes[-1].view_init(elev=90, azim=-90)
    axes[-1].set_xlim([-1.2, 1.2])
    axes[-1].set_ylim([-1.2, 1.2])
    axes[-1].set_zlim([-1.2, 1.2])
    axes[-1].tick_params(axis='both', labelsize=12)
    
    axes[0].axvline(x=64, color='blue', label='Blue Line at x=64',linestyle='--')
    axes[0].axvline(x=32, color='green', label='Green Line at x=32',linestyle='--')

    # Second subplot
    axes[1].axvline(x=16, color='red', label='Red Line at x=16',linestyle='--')
    
    plt.tight_layout()
    plt.subplots_adjust(wspace = 0.5)
#     plt.suptitle(title, fontsize =14)
```

```python
sic = gs.DataFile(datdir + 'SIC_97.dat', x ='x', y = 'y', tmin = -999999)
gs.location_plot(sic, var ='rainfall_x10_mm', cmap ='jet',figsize=(8,4),
                vlim = (np.quantile(sic['rainfall_x10_mm'],0.1),np.quantile(sic['rainfall_x10_mm'],0.9)),
                 plot_style ='mpldefault',title ='', cbar_label = 'Rainfall x10 mm ')
gs.export_image(figdir +'sic_data.png')
```

```python
autovar = gs.Program(exedir+'autovar.exe',getpar = True)
```

```python
parstr = """                Parameters for Autovar
                ************************

START OF PARAMETERS:
{data}SIC_97.dat               -file with data
2  3  0  4  1                  -   columns for X,Y,Z,var,id
-1.0e21   1.0e21               -   trimming limits
0   1                          -Geolocated variogram? (1=yes), Number of anchors
anchor.out                     -file with anchors
2 3 4                          -X Y Z column of anchor file
0                              -Attemp to improve variogram?
0  2  4                        - Use general anisotropy ratio,maj/min,maj/ter
333 -159317.0 1000.0           -nx,xmn,xsiz
429 -108763.0 500.0            -ny,ymn,ysiz
1 0.5 1.0                      -nz,zmn,zsiz
outputs/autovarmod_sic.out     -file for variogram model output
outputs/autovar_sic.out        -   file for experimental pairs output
outputs/debug_sic.out          -   file for directions
outputs/varfit_sic.out         -   variogram model text GSLIB format
outputs/weights_sic.out        -   weights for each anchor
1                              -   Debug on screen (1=yes)
1                              - Disseminated(1) or stratigraphic(2) deposit type? 
"""
autovar.run(parstr =parstr.format(data = datdir),liveoutput = False)
```

```python
# Variograms and directions per anchor
model = gs.DataFile('outputs/autovarmod_sic.out')
exp = gs.DataFile('outputs/autovar_sic.out')
dbg = gs.DataFile('outputs/debug_sic.out')
xlimmaj=150000
xlimmin=150000
xlimter=100
name = 'sic'
plot_autovar2d(model,exp,dbg,name,xlimmaj,xlimmin,xlimter, 'General Variogram')
# Models
varfit = open('outputs/varfit_sic.out' , "r")
varfitc = varfit.read()

print("\033[1m"+'General Variogram model'+"\033[0m")
print(varfitc)
gs.export_image(figdir + 'sic_auto.png', dpi =300)
```

```python
# Plotting function for synthetic datasets
griddef = gs.GridDef('''100   -2.5    1.0                                  -nx,xmn,xsiz
100   -2.5    1.0                                  -ny,ymn,ysiz
100   -2.5    1.0                                  -nz,zmn,zsiz''')
def plot_syn(syn_data):
    fig,axes = plt.subplots(1,2, figsize =(6,3))
    gs.location_plot(syn_data, var = 'value', cmap ='terrain', ax = axes[0],orient = 'xy', edgecolor ='k',title ='')
    gs.location_plot(syn_data, var = 'value', cmap ='terrain', ax = axes[1],orient = 'xz', edgecolor ='k',
                     dhid = 'Drill ID',title ='', griddef = griddef, slicetol = 5, slice_number = 5)
    plt.tight_layout()
    plt.subplots_adjust(wspace = 0.5)
    for ax in [axes[0], axes[1]]:
        ax.tick_params(direction='out', length=3, width=1, colors='k', grid_color='r', grid_alpha=0.5)
        desired_xticks = [10, 30, 50, 70, 90]
        desired_yticks = [10, 30, 50, 70, 90]
        ax.set_xticks(desired_xticks)
        ax.set_yticks(desired_yticks)
```
